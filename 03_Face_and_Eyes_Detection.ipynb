### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Create Classifiers

```python
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
```

### Open Image

```python
image  = cv2.imread ('Images/eye_face.jpg')
```

### Fix Image

```python
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

### Face Classifier

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

### If No Face Detected

```python
if faces is():
    print('No faces found')
```

### It's a Kind of Magic

```python
def detect_face_and_eye (fix_img):
    
    face_rectangles = face_classifier.detectMultiScale(fix_img)
    eye_rectangles = eye_classifier.detectMultiScale(fix_img)
    
     
    for (x, y, w, h) in face_rectangles:
        cv2.rectangle(fix_img, (x, y), (x+w, y+h), (255, 0, 0), 10)
    
    for (x, y, w, h) in eye_rectangles:
        cv2.rectangle(fix_img, (x, y), (x+w, y+h), (0, 0, 255), 5)
        
    return fix_img
```

```python
result = detect_face_and_eye(fix_img)
plt.imshow(result)
```
